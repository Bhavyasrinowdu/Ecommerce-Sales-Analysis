```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/ecommerce_sales_data.csv")
```

```python
df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False)
```

```python
df.groupby("Category")["Sales"].sum()
```

```python
df.groupby("Category")["Sales"].sum().plot(kind="bar")

plt.title("Sales by Category")
plt.show()
```

```python
df.groupby("Region")["Sales"].sum().plot(kind="bar")

plt.title("Sales by Region")
plt.show()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.month

monthly_sales = df.groupby("Month")["Sales"].sum()

monthly_sales.plot(kind="line")

plt.title("Monthly Sales Trend")

plt.show()
```

```python
top_products = df.groupby("Product Name")["Sales"].sum()

top_products.sort_values(ascending=False).plot(kind="bar")

plt.title("Sales by Product")
plt.show()
```
